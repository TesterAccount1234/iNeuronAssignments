ðŸ’¡ Question-1:

Given a Binary Tree (Bt), convert it to a Doubly Linked List(DLL). The left and right pointers in nodes are to be used as previous and next pointers respectively in converted DLL. The order of nodes in DLL must be the same as in Inorder for the given Binary Tree. The first node of Inorder traversal (leftmost node in BT) must be the head node of the DLL.

</aside>

```python
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

    
def convertToDLL(root):
    if root is None:
        return None
    
    # Convert left subtree to DLL
    if root.left:
        left_head = convertToDLL(root.left)
        left_tail = left_head.left
        
        left_tail.right = root
        root.left = left_tail
        root_head = left_head
    else:
        root_head = root

    # Convert right subtree to DLL
    if root.right:
        right_head = convertToDLL(root.right)
        right_tail = right_head.left
        
        root.right = right_head
        right_head.left = root
        root_tail = right_tail
    else:
        root_tail = root

    root_head.left = root_tail
    root_tail.right = root_head
    
    return root_head
    

def printDLL(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.right
        if current == head:
            break

root = Node(10)
root.left = Node(5)
root.right = Node(20)
root.right.left = Node(30)
root.right.right = Node(35)

head = convertToDLL(root)

printDLL(head)
```

Given a binary tree, print all its root-to-leaf paths without using recursion. For example, consider the following Binary Tree.

Input:

        6
     /    \
    3      5
  /   \     \
 2     5     4
     /   \
    7     4

Output:

There are 4 leaves, hence 4 root to leaf paths -
  6->3->2
  6->3->5->7
  6->3->5->4
  6->5>4

</aside>

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def pathToLeaf(root):
    if root is None:
        return 
    
    # Pos, value
    stack = [(root, str(root.val))]

    while stack:
        node, path = stack.pop()

        #Leaf node:
        if node.left == None and node.right == None:
            print(path)

        if node.right:
            stack.append((node.right, path + '->' + str(node.right.val)))
        if node.left:
            stack.append((node.left, path + '->' + str(node.left.val)))


root = TreeNode(6)
root.left = TreeNode(3)
root.right = TreeNode(5)
root.left.left = TreeNode(2)
root.left.right = TreeNode(5)
root.right.right = TreeNode(4)
root.left.right.left = TreeNode(7)
root.left.right.right = TreeNode(4)

pathToLeaf(root)
```

GivenÂ Preorder, Inorder and PostorderÂ traversals of some tree. Write a program to check if they all are of the same tree.

**Examples:**

Input : 

        Inorder -> 4 2 5 1 3
        Preorder -> 1 2 4 5 3
        Postorder -> 4 5 2 3 1
Output : 

Yes
Explanation : 

All of the above three traversals are of
the same tree 

                           1
                         /   \
                        2     3
                      /   \
                     4     5

Input : 

        Inorder -> 4 2 5 1 3
        Preorder -> 1 5 4 2 3
        Postorder -> 4 1 2 3 5
Output : 

No

</aside>

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def checkSameTree(preorder, inorder, postorder):
    #Empty
    if not preorder or not inorder or not postorder:
        return False
    
    if len(preorder) != len(inorder) or len(preorder) != len(postorder):
        return False

    if len(preorder) == len(inorder) == len(postorder) == 1:
        return preorder[0] == inorder[0] == postorder[0]

    root = preorder[0]
    root_index = inorder.index(root)

    if inorder[:root_index] != preorder[1:root_index + 1]:
        return False

    if inorder[root_index + 1:] != postorder[:len(inorder) - root_index - 1]:
        return False

    left_subtree = checkSameTree(preorder[1:root_index + 1], inorder[:root_index], postorder[:root_index])
    right_subtree = checkSameTree(preorder[root_index + 1:], inorder[root_index + 1:], postorder[root_index:len(postorder) - 1])

    return left_subtree and right_subtree

inorder = [4, 2, 5, 1, 3]
preorder = [1, 2, 4, 5, 3]
postorder = [4, 5, 2, 3, 1]

checkSameTree(preorder, inorder, postorder)
```
